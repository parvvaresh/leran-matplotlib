```python
import matplotlib.pyplot as pt
x1 = [1,3,4,5,6]
y1 = [1,2,3,67,5]

x2 = [1,2,7,8,11]
y2 = [1,20,37,47,50]

x3 = [1,2,3,4,5]
y3 = [22,33,44,55,66]

pt.plot(x1,y1,linestyle="--",linewidth=3,color="red")
pt.plot(x2,y2,linestyle="--",linewidth=3,color="blue")
pt.plot(x3,y3,linewidth=3,color="black")

pt.xlabel("X")
pt.ylabel("Y")
pt.show()
```

```python
import matplotlib.pyplot as pt
x1 = [1,3,4,5,6]
y1 = [1,2,3,67,5]

x2 = [1,2,7,8,11]
y2 = [1,20,37,47,50]

x3 = [1,2,3,4,5]
y3 = [22,33,44,55,66]

pt.plot(x1,y1,linestyle="--",linewidth=3,color="red",label="line1")
pt.plot(x2,y2,linestyle="--",linewidth=3,color="blue",label="line1")
pt.plot(x3,y3,linewidth=3,color="black",label="line1")

pt.xlabel("X")
pt.ylabel("Y")
pt.legend()
pt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pt
from matplotlib.pyplot import style

from random import choice
number_subplot= 25

for i in range(number_subplot):
  value = int(number_subplot ** 0.5)
  x,y = choice(range(300)),choice(range(300))
  pt.subplot(value, value, i+1)
  pt.plot(x,y)
pt.show()
```

```python
import numpy as np
import matplotlib.pyplot as pt

from matplotlib.pyplot import style

style.use('ggplot')

days = [1,2,3,4,5,6,7]
code = [3,4,5,0,4,9,1]
sleap = [10,5,9,11,12,4,12]
eat = [1,1,1,2,2,1,1]

stack = np.vstack([code, sleap, eat])

print(stack, type(stack))

pt.stackplot(days, stack)
pt.show()
```
